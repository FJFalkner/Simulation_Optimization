# Sky Diver




```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from explicit_integration import *


# ODE of sky diver
def dydt(t, y):
    
    g = 9.81        # gravity acceleration / m/s^2
    m = 80          # mass of sky diver / kg
    k1 = 0.33       # drag coefficient before opening parachute / kg/m
    k2 = 33         # drag coefficient after opening parachute / kg/m
    t1 = 30.05      # time of opening parachute / s
    t2 = 34         # time of fully opened parachute / s

    # free fall
    if t < t1:
        f = g - k1/m * y**2
        
    # opening parachute
    elif t1 <= t < t2:
        f = g - (k2-k1)/(t2 - t1)*(t - t1)/m * y**2
    
    # landing
    else:
        f = g - k2/m * y**2

    return f
```

Solution

```python
# time integration
# n_crit = 86
t, y = implicit_Euler(dydt, (0, 50), [0], 30)

# plot result
plt.plot(t, y,'-o', markersize=3)
plt.xlabel('t / s')
plt.ylabel('v /ms$^{-1}$')
plt.grid()
plt.title('Sky Diver')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```

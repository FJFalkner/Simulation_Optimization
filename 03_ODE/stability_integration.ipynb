# Stability of Numerical Time Integration

```python
import numpy as np
import matplotlib.pyplot as plt
from explicit_integration import *
```

Solve the IVP

$$
y' = ay \qquad \mathrm{and} \qquad y(0) = 1
$$

```python
# time span
t_span = (0, 1)

# inital conditions
y0 = [1]

# number of steps
n = 51

# ode
a = -100
def dydt(t, y):
    a = -100
    return a*y

# solution
t, y = explicit_Euler(dydt, t_span, y0, n)
```

Plot results

```python
plt.plot(t,y)
plt.plot(t, np.exp(a*t), 'k--')
plt.grid()
plt.xlabel('t / s')
plt.ylabel('y')
plt.show()
```

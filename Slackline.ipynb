Import Libraries and define system parameters

```python
import numpy as np
import matplotlib.pyplot as plt

l = 4       # span in m
T0 = 1      # pre-tension force in kN
EA = 80     # axial stiffness in kN
G = 0.8     # applied force in kN
```

### Nonlinear equation

Equlibrium in the *undeformed* configuration in the vertical direction yields
$$
f(x) = 
$$

```python
def f(x, l=l, T0=T0, G=G, EA=EA):
    return 2 * (EA * (np.sqrt(1 + 4 * x**2 / l**2) - 1) + T0) * x / np.sqrt(l**2 / 4 + x**2) - G
```

Plot nonlinear equation

```python
x_max = 0.8
x_plot = np.linspace(0, x_max, 100)
f_plot = f(x_plot)

plt.plot(x_plot, f_plot)
plt.grid()
plt.xlabel('x / m')
plt.ylabel('f(x) / kN')
plt.xlim(0, x_max)
plt.title('Nonlinear equation of slackline')
plt.show()
```

## Bisection method

```python
# intial interval
xl = 0.1
xr = 0.8
# tolerance and maximum number of iterations
tol = 1E-6
max_iter = 50

# first iteration
xm = 0.5*(xr + xl)
fm = f(xm)
fl = f(xl)
fr = f(xr)

# start bisection method
for i in range(0, max_iter):

    # new interval
    if fm*fl < 0:
        xr = xm
        fr = fm
    else:
        xl = xm
        fl = fm

    # new midpoint and function evaluation
    xm = 0.5*(xr + xl)
    fm = f(xm)

    # output
    print(f"{i:3d}   {xm:6.5f}   {fm:+5.4E}")

    # convergence check
    if abs(fm) < tol:
        break
```
